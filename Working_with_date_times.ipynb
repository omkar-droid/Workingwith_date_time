```python
#converting DATE to datetime datatype

date['date']=pd.to_datetime(date['date'])
```

```python
date.info()
```

```python
# extract year

date['date_year'] = date['date'].dt.year

date.head()
```

```python
#extract MONTH

date['date_month'] = date['date'].dt.month
date.head()
```

```python
#extract  MONTH NAME


date['date_month_name'] = date['date'].dt.month_name()
date.head()
```

```python
#extract  DAY


date['date_day'] = date['date'].dt.day
date.head()
```

```python
#extract  DAY


date['date_day'] = date['date'].dt.day
date.head()
```

```python
#extract  DAY of week


date['date_day_week'] = date['date'].dt.day_of_week
date.head()
```

```python
#extract  DAY of year


date['date_day_year'] = date['date'].dt.day_of_year
date.head()
```

```python
#extract  DAY of year


date['date_day_name'] = date['date'].dt.day_name()
date.head()
```

```python
# is Weekend

date['date_is_weekend'] = np.where(date['date_day_name'].isin(['Saturday','Sunday']),1,0)

date.drop(columns=['product_id','city_id','orders']).head()
```

```python
# Extract WEEK

date['week'] = date['date'].dt.isocalendar().week

date.drop(columns=['product_id','city_id', 'orders']).head()

```

```python
# extract quarter

date['quarter'] = date['date'].dt.quarter
date.head()
```

```python
# extraxt semester


date['semester'] = np.where(date['quarter'].isin([1,2]),1,2)
date.drop(columns=['product_id','city_id', 'orders']).head()
```

```python
#extract TIME Elapsed between dates




import datetime

today = datetime.datetime.today()

today
```

```python
today - date['date']
```

```python
(today - date['date']).dt.days
```

```python
# Months PASSED
from dateutil.relativedelta import relativedelta
# Calculate months passed using relativedelta
date['months_passed'] = date['date'].apply(lambda x: relativedelta(today, x).months + relativedelta(today, x).years * 12).head()
date.head()
```
